# 4. Find the top 5 cars based on their sales price.

```python
a = data.nlargest(5,["Sale Price"])
print(a[["Make","Model","Nickname","Sale Price"]])
```

# 5. Find the least 5 cars based on their Resell price.

```python
a = data.nsmallest(5,["Resell Price"])
print(a[["Make","Model","Nickname","Resell Price"]])
```
